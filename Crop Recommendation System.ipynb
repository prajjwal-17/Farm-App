```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('Crop_recommendation.csv')
df.head(20)
```

```python
df.isnull().sum()
```

```python
df['Crops'].value_counts()
```

```python
x=df.drop('Crops', axis=1)
y = df['Crops']
```

```python
x.info()
```

```python
y.info()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state =1, test_size = 0.2)
```

```python
x_train.info()
```

```python
x_test.info()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
y_pred1 = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
logistic_reg_acc = accuracy_score(y_test, y_pred1)
print("logistic accuracy is "+ str(logistic_reg_acc))
```

```python
from sklearn.tree import DecisionTreeClassifier
model2 = DecisionTreeClassifier()
model2.fit(x_train, y_train)
y_pred3 = model2.predict(x_test)
```

```python
decision_acc = accuracy_score(y_test, y_pred3)
print("decision accuracy is "+ str(decision_acc))
```

```python
from sklearn.ensemble import RandomForestClassifier
model3 = RandomForestClassifier()
model3.fit(x_train, y_train)
y_pred4 = model3.predict(x_test)
```

```python
random_acc = accuracy_score(y_test, y_pred4)
print("Random forest accuracy is "+ str(random_acc))
```

```python
import joblib

# Save the trained model
joblib.dump(model2, "crop_recommendation_model.pkl")
print("Model saved successfully!")
```

```python
# Load the trained model
loaded_model = joblib.load("crop_recommendation_model.pkl")
print("Model loaded successfully!")
```

```python
# Example input (Ensure it has the correct shape)
arr = [[90, 42, 43, 20.879744, 82.002744, 6.502985, 202.935536]]  # Modify as needed

# Predict using the loaded model
y_pred5 = loaded_model.predict(arr)
print("Predicted Crop:", y_pred5[0])
```
